```python
import pandas as pd
videogamesales = pd.read_csv("./input/vgsales.csv")
```

```python
contingencyTable = videogamesales.groupby(['Platform','Genre']).size().unstack('Genre', fill_value=0)
contingencyTable
```

```python
contingencyTable.plot(kind='bar')
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(contingencyTable)
```

Er is geen sprake van een verband tussen genre en platform

```python
contingencyTable2 = videogamesales.groupby(['Platform','Publisher']).size().unstack('Publisher', fill_value=0)
contingencyTable2
```

```python
contingencyTable2.plot(kind='bar')
```

```python
chi2_contingency(contingencyTable2)
```

Er is geen sprake van een verband tussen publisher en platform
